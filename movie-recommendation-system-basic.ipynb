```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
movies = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0]['genres']
```

```python
import ast
def convert(text):
    h = []
    for i in ast.literal_eval(text):
        h.append(i['name']) 
    return h
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0]['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies.iloc[0]['cast']
```

```python
def convert_cast(text):
    c = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            c.append(i['name'])
        counter+=1
    return c
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head()
```

```python
movies.iloc[0]['crew']
```

```python
def pull_director(text):
    D = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            D.append(i['name'])
            break
    return D
```

```python
movies['crew'] = movies['crew'].apply(pull_director)
movies.head()
```

```python
movies.iloc[0]['overview']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies.iloc[0]['overview']
```

```python
def remove_space(L):
    S = []
    for i in L:
        S.append(i.replace(" ",""))
    return S
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies.iloc[0]['tags']
```

```python
Movie_df = movies[['movie_id','title','tags']]
```

```python
Movie_df.head()
```

```python
Movie_df['tags'] = Movie_df['tags'].apply(lambda x: " ".join(x))
Movie_df.head()
```

```python
Movie_df.iloc[0]['tags']
```

```python
Movie_df['tags'] = Movie_df['tags'].apply(lambda x:x.lower())
```

```python
print(Movie_df.head)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# TF-IDF vektörlerini oluşturun
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(Movie_df['tags'].values.astype('U'))  # 'tags' sütununu kullanarak vektörleri oluşturuyoruz
```

```python
# Kosinüs benzerliğini hesaplayın
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # Film başlığına göre indeksi alın
    idx = Movie_df[Movie_df['title'] == title].index[0]

    # Tüm filmlerin benzerlik skorlarını alın
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Benzerlik skorlarına göre filmleri sıralayın
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # En iyi 10 filmi alın
    sim_scores = sim_scores[1:11]

    # Film indekslerini alın
    movie_indices = [i[0] for i in sim_scores]

    # İlgili film başlıklarını döndürün
    return Movie_df['title'].iloc[movie_indices]
```

```python
# Önerileri alın
film_basligi = "Interstellar"  # Kullanıcı tarafından seçilen bir film başlığı
print("Film Önerileri:")
print(get_recommendations(film_basligi))
```
